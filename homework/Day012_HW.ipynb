### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date 
import re

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
day_today = date.today()
day__today = str(day_today).replace('-','/')
print(day__today)
```

```python
category = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    title = d.find('a').text
    retitle= str(title).replace('\u3000',' ')
    category.append(d.find_all('em')[-1].text)
    category.append(retitle)
print(category)
```

### ③ 哪一個類別的文章最多

```python
category1 = []

for k in soup.find(class_="part_list_2").find_all('h3'):
    category1.append(k.find_all('em')[-1].text)
```

```python
from collections import Counter
import pandas as pd
category1.sort()
result = Counter(category1)
result2 = pd.value_counts(category1)
print(result)
print(result2)
```
